```python
import numpy as np
import scipy.stats as stats
import xarray as xr
import netCDF4 as nc
from pathlib import Path

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')
```

```python
def ax_pos_inch_to_absolute(fig_size, ax_pos_inch):
    ax_pos_absolute = []
    ax_pos_absolute.append(ax_pos_inch[0]/fig_size[0])
    ax_pos_absolute.append(ax_pos_inch[1]/fig_size[1])
    ax_pos_absolute.append(ax_pos_inch[2]/fig_size[0])
    ax_pos_absolute.append(ax_pos_inch[3]/fig_size[1])

    return ax_pos_absolute
```

```python
def ax_pos_cm_to_absolute(fig_size, ax_pos_cm):
    ax_pos_absolute = []
    ax_pos_inch = [ pos / 2.54 for pos in ax_pos_cm ]
    ax_pos_absolute.append(ax_pos_inch[0]/fig_size[0])
    ax_pos_absolute.append(ax_pos_inch[1]/fig_size[1])
    ax_pos_absolute.append(ax_pos_inch[2]/fig_size[0])
    ax_pos_absolute.append(ax_pos_inch[3]/fig_size[1])
    
    return ax_pos_absolute
```

```python
# base dir
base_dir = (Path.cwd() / "../../").resolve()
data_dir = base_dir / "data"
save_dir = base_dir / "figures"
```

```python
file_name = "boot-statistics-grid-space-realization-2024-epsilon0-5.8-lambda0-0.06-tau0-2.3.nc"
ds_boot = xr.open_dataset(str(data_dir / file_name))
```

```python
# extract
statistic = ds_boot.statistic.values
boot_statistic = ds_boot.boot_statistic.values
p_value = ds_boot.p_value.values

lond = ds_boot.long.values
latd = ds_boot.latg.values
```

```python
# indices to plot
ind_lon = [np.argmin(np.abs(lond -  40)), np.argmin(np.abs(lond -  80)), np.argmin(np.abs(lond - 120)),
           np.argmin(np.abs(lond - 160)), np.argmin(np.abs(lond - 200)), np.argmin(np.abs(lond - 240)),
           np.argmin(np.abs(lond - 280)), np.argmin(np.abs(lond - 320)), np.argmin(np.abs(lond - 360))]

ind_lat = [np.argmin(np.abs(latd - 30)), np.argmin(np.abs(latd - 30)), np.argmin(np.abs(latd - 30)),
           np.argmin(np.abs(latd - 30)), np.argmin(np.abs(latd - 30)), np.argmin(np.abs(latd - 30)),
           np.argmin(np.abs(latd - 30)), np.argmin(np.abs(latd - 30)), np.argmin(np.abs(latd - 30))]
```

```python
# plot bootstrap statistic distribution

fig_size = (17.40/2.54, 16.75/2.54)
fig = plt.figure(figsize=fig_size)

ax = []
lax = []

ax.append(fig.add_axes(ax_pos_cm_to_absolute(fig_size, [00.50, 12.00, 04.00, 04.00])))
ax.append(fig.add_axes(ax_pos_cm_to_absolute(fig_size, [05.25, 12.00, 04.00, 04.00])))
ax.append(fig.add_axes(ax_pos_cm_to_absolute(fig_size, [10.00, 12.00, 04.00, 04.00])))

ax.append(fig.add_axes(ax_pos_cm_to_absolute(fig_size, [00.50, 06.75, 04.00, 04.00])))
ax.append(fig.add_axes(ax_pos_cm_to_absolute(fig_size, [05.25, 06.75, 04.00, 04.00])))
ax.append(fig.add_axes(ax_pos_cm_to_absolute(fig_size, [10.00, 06.75, 04.00, 04.00])))

ax.append(fig.add_axes(ax_pos_cm_to_absolute(fig_size, [00.50, 01.50, 04.00, 04.00])))
ax.append(fig.add_axes(ax_pos_cm_to_absolute(fig_size, [05.25, 01.50, 04.00, 04.00])))
ax.append(fig.add_axes(ax_pos_cm_to_absolute(fig_size, [10.00, 01.50, 04.00, 04.00])))

lax.append(fig.add_axes(ax_pos_cm_to_absolute(fig_size, [14.20, 01.50, 03.00, 12.50])))

panel_list = ['A', 'B', 'C',
              'D', 'E', 'F',
              'G', 'H', 'I',
              'J', 'K', 'L',
              'M', 'N', 'O',
              'P', 'Q', 'R']

edges = [[0.7, 1.3], [0.7, 1.3], [0.7, 1.3],
         [0.7, 1.3], [0.7, 1.3], [0.7, 1.3],
         [0.7, 1.3], [0.7, 1.3], [0.7, 1.3]]

dofs = [  900, 2000, 1400,
         1900, 1400,  600,
         1400, 1400,  800]

xlims = [[0.7, 1.3], [0.7, 1.3], [0.7, 1.3],
         [0.7, 1.3], [0.7, 1.3], [0.7, 1.3],
         [0.7, 1.3], [0.7, 1.3], [0.7, 1.3]]

### ax0-ax8 ###

for i in range(9):

    # ax[i].grid()

    bins = np.histogram_bin_edges(edges[i], bins=20)

    p0 = sns.histplot(boot_statistic[:, ind_lat[i], ind_lon[i]], bins=bins, color='C1', edgecolor='white', ax=ax[i], alpha=0.5, stat='density', label=r'Bootstrap')

    # color bars greater than p-value
    if (statistic[ind_lat[i], ind_lon[i]] > 1):
        for rectangle in p0.patches:
            if rectangle.get_x() + 0.5*rectangle.get_width() >= statistic[ind_lat[i], ind_lon[i]]:
                rectangle.set_facecolor('C1')
            if rectangle.get_x() + 0.5*rectangle.get_width() <= 1/statistic[ind_lat[i], ind_lon[i]]:
                rectangle.set_facecolor('C1')
    if (statistic[ind_lat[i], ind_lon[i]] < 1):
        for rectangle in p0.patches:
            if rectangle.get_x() <= statistic[ind_lat[i], ind_lon[i]]:
                rectangle.set_facecolor('C1')
            if rectangle.get_x() >= 1/statistic[ind_lat[i], ind_lon[i]]:
                rectangle.set_facecolor('C1')

    x = np.linspace(edges[i][0], edges[i][1], 100)
    f_dist = stats.f.pdf(x, dofs[i], dofs[i])
    ax[i].plot(x, f_dist, color='black', label='F-dist')

    ax[i].axvline(x=statistic[ind_lat[i], ind_lon[i]], ymin=0, ymax=1, color='C1', linewidth=1)
    ax[i].axvline(x=1/statistic[ind_lat[i], ind_lon[i]], ymin=0, ymax=1, color='C1', linewidth=1)

    ax[i].set_xlim(xlims[i])
    ax[i].ticklabel_format(style='sci', axis='y', scilimits=(0,0))
    ax[i].yaxis.offsetText.set(ha="right")
    ax[i].yaxis.offsetText.set_position((0, 0))

    if i in [6, 7, 8]:
        ax[i].set_xlabel(r'$S_{1}^{2}$ / $S_{2}^{2}$')
        # ax[i].xaxis.set_label_coords(0.5, -0.1)

    ax[i].set_ylabel('')

    if (lond[ind_lon[i]] > 180) and (latd[ind_lat[i]] < 0):
        ax[i].set_title(r'({:.0f}°W,{:.0f}°S)'.format(360 - lond[ind_lon[i]], latd[ind_lat[i]]), fontsize=10, loc='center')
    elif (lond[ind_lon[i]] > 180) and (latd[ind_lat[i]] > 0):
        ax[i].set_title(r'({:.0f}°W,{:.0f}°N)'.format(360 - lond[ind_lon[i]], latd[ind_lat[i]]), fontsize=10, loc='center')
    elif (lond[ind_lon[i]] > 0) and (latd[ind_lat[i]] < 0):
        ax[i].set_title(r'({:.0f}°E,{:.0f}°S)'.format(lond[ind_lon[i]], latd[ind_lat[i]]), fontsize=10, loc='center')
    elif (lond[ind_lon[i]] > 0) and (latd[ind_lat[i]] > 0):
        ax[i].set_title(r'({:.0f}°E,{:.0f}°N)'.format(lond[ind_lon[i]], latd[ind_lat[i]]), fontsize=10, loc='center')

h0, l0 = ax[0].get_legend_handles_labels()
# h6, l6 = ax[0].get_legend_handles_labels()
# h9, l9 = ax[0].get_legend_handles_labels()
ax[2].legend(h0, l0, bbox_to_anchor=(1.85, 1), loc='upper right', fontsize=10)

for i in range(9):
    ax[i].text(00.05, 00.95, panel_list[i], ha='left', va='top', transform=ax[i].transAxes, fontsize=10, fontweight='bold', color='black')

# legend
lax[0].set_axis_off()

for i in range(9):
    lax[0].text(00.05, 00.99-i*0.11, panel_list[i], ha='left', va='top', transform=lax[0].transAxes, fontsize=10, fontweight='bold', color='black')
    lax[0].text(00.05, 00.95-i*0.11, "DOFs: ({:.0f}, {:.0f})\np-value: {:.3f}".format(dofs[i], dofs[i], p_value[ind_lat[i], ind_lon[i]]), ha='left', va='top', transform=lax[0].transAxes, fontsize=8, color='black', backgroundcolor='white')
```

```python
file_name = "fig-s02"
Path(save_dir).mkdir(parents=True, exist_ok=True)
fig.savefig(str(save_dir / file_name) + ".png", dpi=600, format='png', facecolor='white')
fig.savefig(str(save_dir / file_name) + ".pdf", dpi=600, format='pdf', facecolor='white')
```
